```python
class Numero:
    data = ['1','2','3','4']
    rep = 1234        
    def __init__(self,num):
        self.data[0] = str(int(num/1000))
        self.data[1] = str(int((num%1000)/100))
        self.data[2] = str(int((num%100)/10))
        self.data[3] = str(int(num%10))
        self.rep = num
    def normalize(self):
        self.rep = int(self.data[0])*1000 + int(self.data[1])*100 + int(self.data[2])*10 + int(self.data[3])
    def show(self):
        if(self.rep<1000):
            print("0",end="")
        print(self.rep)
    def get(self, index):
        return self.data[index]
    def contains(self,num):
        aux = False
        for i in self.data:
            if(i == num):
                aux = True
        return aux
    def replace(self,index,num):
        self.data[index] = num
        self.normalize()
    def mix(self,index):
        aux = ['X']
        j = 0
        for i in index:
            if(i):
                aux.append(self.data[j])
            j+=1
        aux[0]=aux[len(aux)-1]
        aux.pop()
        j = 0
        for i in index:
            if(i):
                self.data[j]=aux[0]
                aux.remove(self.data[j])
            j+=1
        self.normalize()
    def toInt(self):
        return self.rep
    
num = Numero(5679)
num.show()
num.replace(2,'0')
num.show()
num.mix([1,0,1,1])
num.show()
num.mix([1,0,1,1])
num.show()
num.mix([1,0,1,1])
num.show()
```
